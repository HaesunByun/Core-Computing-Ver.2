### Problem: (스스로) 원본 데이터를 보호하라!
😄 회사의 매출 데이터를 관리하는 프로그램을 작성하는 건혁이는 각 분기별 매출 데이터를 복사하여 각 지점에게 전달했습니다. 그러나 지점들이 데이터를 수정하는 과정에서 원본 데이터가 의도치 않게 변경되는 문제가 발생하고 말았습니다. 원본 데이터의 변경 없이 1분기 인천지점의 매출액을 1350, 4분기 서울지점의 매출액을 4320으로 수정하시오.

---

<font color='blue'>`Test case-1`</font>

Output Sample
```
Original Sales: [[1000, 2000, 3000], [1500, 2500, 3500], [2000, 3000, 4000], [2500, 3500, 4500]]
인천지점 Sales: [[1350, 2000, 3000], [1500, 2500, 3500], [2000, 3000, 4000], [2500, 3500, 4500]]
서울지점 Sales: [[1000, 2000, 3000], [1500, 2500, 3500], [2000, 3000, 4000], [2500, 3500, 4320]]
```

````python
import copy

# 1. 원본 데이터 생성
original_sales = [
    [1000, 2000, 3000],  # 1분기 매출 (인천지점, 경기지점, 서울지점)
    [1500, 2500, 3500],  # 2분기 매출
    [2000, 3000, 4000],  # 3분기 매출
    [2500, 3500, 4500],  # 4분기 매출
]

copy1_sales = copy.copy(original_sales)
copy1_sales[0][0] = 1350

print("Original Sales:", original_sales)
print("인천지점 Sales:", copy1_sales)
````
