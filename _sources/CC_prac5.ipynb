### Problem: (함께) Counter 클래스◼◼◻◻◻
😄 Counter는 기계식 계수기이며 경기장이나 콘서트에 입장하는 관객 수를 세기 위하여 사용합니다. Counter 클래스를 정의하시오.

<div align="center"><img src="https://haesunbyun.github.io/common/images/class1.png" style="width:300px;"></div>

- Counter는 다음과 같은 속성과 동작을 가집니다.
  - 관객 수를 세기 위한 count 변수와 객체를 구분하기 위한 이름을 가진다.
  - `__init__()`: Counter 클래스를 초기화한다.
  - reset(): count 변수를 0으로 초기화한다.
  - increment(): count 변수를 1 증가한다.
  - get(): count 값을 반환한다.



<font color='blue'>`test case-1`</font>

Sample Code
```python
a = Counter('counter1')
b = Counter('counter2')
for i in range(10):
  a.increment()
  b.increment()
print(a.get())
print(b.get())
a.reset()
print(a.get())
```

Output Sample
```
10
10
counter1이 초기화 되었습니다.
0
```

### Problem: (함께+스스로)시저암호◼◼◼◼◻

😄 어떤 문장의 각 알파벳을 일정한 거리만큼 밀어서 다른 알파벳으로 바꾸는 암호화 방식을 시저 암호라고 합니다. 예를 들어 "AB"는 1만큼 밀면 "BC"가 되고, 3만큼 밀면 "DE"가 됩니다. "z"는 1만큼 밀면 "a"가 됩니다.

Sender의 기능은 두 가지입니다.
  - slider()는 n과 m를 전달받아 n만큼 민 시저 암호를 만들어 리턴합니다.

  - decode()는 n만 전달받아 저장한 암호를 해독하여 출력합니다.

### Problem: Box 클래스◼◼◼◼◼
😄 직사각형을 클래스로 나타낸 뒤, 직사각형끼리 연산해봅니다.
구현해야 할 연산은 세 가지 입니다.
  - intersect: 두 개의 직사각형이 겹치는 구간의 원점좌표 리스트와 높이, 너비를 Box 타입으로 반환합니다.
  - coverage: 두 개의 직사각형이 커버하는 구간의 원점좌표 리스트와 높이, 너비를 Box 타입으로 반환합니다.
  - to_string: Box 타입의 인스턴스가 출력되는 형식을 지정합니다. to_string()함수로 만들어도 되고, __str__() 함수로 정의해도 됩니다.

입력 형식은 다음과 같습니다. 좌표 상에서 사각형의 맨 왼쪽 아래 점을 '원점' origin으로 합니다. 그리고 그 지점에서 세로 변 길이를 '높이' height, 가로 변 길이를 '너비' width라고 합니다. 한 줄에 입력은 원점 높이 너비 순으로 들어오고, 각각 공백 한 개로 구분되어 있습니다.

원점은 중심점의 x값과 y값을 리스트 형태로 구성한 것입니다. 클래스 내에서도 리스트로 저장해야 합니다. 주어진 직사각형 간의 intersect는 항상 존재합니다. 두 개가 겹쳐지지 않는 경우는 없습니다.

<div align="center"><img src="https://haesunbyun.github.io/common/images/class3.png" style="width:200px;"></div>

- intersection 영역 계산하는 방법

```
(x1,y1) = (max(x1), max(y1))
(x2,y2) = (min(x2), min(y2))
width = x2-x1
height = y2-y1
```
