### Problem: (함께) 우박수열

😄 우박수열은 숫자가 커졌다 작아졌다를 반복하다 결국 1에 수렴하는 걸 비구름에서 빗방울이 오르락내리락하며 우박이 되는 모습에 빗대어 부른 수열이다.

우박수열은 모든 자연수 n에 대해 아래의 수식을 적용하여 유한번 재귀 반복하면 1로 간다는 것으로 수학자 콜라츠가 제기한 추측이다.
$ T(n) = \begin{cases} \frac{n}{2}, \;if \;n \;is \;even \\
3n+1, \;if \;n \;is \;odd \end{cases} $

이 수식은 n이 짝수라면 2로 나누고, 홀수라면 3을 곱하고 1을 더한다. 이 작업을 반복하면 1로 도달한다는 것이다.
예를 들어, 5는 5 → 16 → 8 → 4 → 2 → 1이 된다.
6은 6 → 3 → 10 → 5 → 16 → 8 → 4 → 2 → 1이 된다.

입력받은 수에 대해 우박수열을 구하고 몇 번 만에 1에 도달하게 되는지 출력하는 재귀함수를 정의하시오.

### Problem: 오륜기 그리기
😄 함수 drawCircle(point)는 원이 그려질 x,y좌표를 전달받아 크기가 50인 원을 그리는 함수이다. x,y 좌표값은 다음과 같다.
pointList = [(-80, 0), (0, 0), (80, 0), (-40, -70), (40, -70)]
다음과 같이 오륜기를 그리시오.





---



````python
import ColabTurtlePlus.Turtle as tpt

tpt.clearscreen()
tpt.initializeTurtle()


def drawCircle(point):
  global color_idx
  tpt.up()
  tpt.goto(point)
  tpt.down()
  tpt.color(colors[color_idx])
  color_idx += 1
  tpt.circle(50)

def circle_recursive(pointList):
  if len(pointList)==0:
    return
  else:
    drawCircle(pointList[0])
    circle_recursive(pointList[1:])

colors = [ 'blue', 'black', 'red', 'yellow', 'green']
color_idx=0
tpt.bgcolor("white")
pointList = [(-80, 0), (0, 0), (80, 0), (-40, -70), (40, -70)]
circle_recursive(pointList)
````

### Problem: 벚꽃나무 그리기
😄 가지의 길이(length)와 가지뻗기 단계(level)을 입력받아 나무를 그리시오.
함수 maju( )는 함수 branch()를 호출한다. 가지의 굵기(펜의 굵기)는 level로 설정하는데 level이 1보다 클때까지는 가지뻗기를 하며, 1보다 작으면 분홍잎을 찍는다. maju()함수에 정의된 것과 같이 길이와 각도가 주어진다.

````python
import ColabTurtlePlus.Turtle as t

t.clearscreen()
t.initializeTurtle()
t.clear()
t.home()

def branch(length, level, angle):
  t.pensize(level)    # 굵기 설정
  t.left(angle)		# 각도 변경
  t.forward(length)   # 가지 뻗기
  if level > 1:
      maju(length, level)
  else:
      t.dot(30-inLevel*2, 'pink')	# 점 찍기
  t.back(length)		# 가지 뻗기 전 위치로
  t.right(angle)		# 각도 원위치

def maju(length, level):
  branch(length*0.8, level-1, -27)	# branch 호출
  branch(length*0.7, level-1, 47) 	# branch 호출


inLength = int(input())  #test : inLenth 100
inLevel = int(input())   #test : inLevle 10

t.speed(8)		    # 속도 설정
t.hideturtle()		# 숨기기
t.pensize(inLevel)	# 굵기 설정|
t.left(90)			# 각도 변경
t.back(inLength)	# 나무 줄기 그리기
t.forward(inLength)	# 가지 그리기 시작 위치로 이동

maju(inLength, inLevel)
````
