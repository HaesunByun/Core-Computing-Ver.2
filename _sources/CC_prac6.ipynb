## 스택 구조
<div align="center"><img src="https://haesunbyun.github.io/common/images/adt1.png" width="300" height="200"style="width:600px;"></div>

````python
class Stack:

  def __init__(self):
    '''
    새로운 Stack을 생성한다.
    '''
    self.top = -1
    self.sList = []

  def __str__(self):
    '''
    Stack의 출력 형식을 S<item1, item2, item3, ...>S과 같이 출력한다.
    '''
    value=str(self.sList)[1:-1]
    return "S<"+value+">S"

  def push(self, item):
    '''
    Stack에 item을 넣는다.
    '''
    self.sList.append(item)
    self.top += 1

  def isEmpty(self):
    '''
    Stack에 item이 없으면 True, 있으면 False를 반환한다.
    '''
    if self.top == -1:
      return True
    return False

  def pop(self):
    '''
    Stack에 item이 있으면 item을 반환하고 Stack에서는 제거한다.
    Stack이 비어 있으면 'Stack is Empty'를 반환한다.
    '''
    if not(self.isEmpty()):
      self.top -= 1
      return self.sList.pop()
    return 'Stack is Empty'

  def peek(self):
    '''
    맨 위의 아이템을 반환한다.
    '''
    if not(self.isEmpty()):
      return self.sList[-1]

  def size(self):
    return len(self.sList)
````

## 큐 구조

````python
class Queue:

  def __init__(self):
    self.qList = []

  def __str__(self):
    value=str(self.qList)[1:-1]
    return f"Q<{value}>Q"

  def isEmpty(self):
    return not bool(self.qList)

  def enqueue(self, item):
    self.qList.append(item)

  def dequeue(self):
    if not(self.isEmpty()):
      return self.qList.pop(0)

  def peek(self):
    if not(self.isEmpty()):
      return self.qList[0]

  def size(self):
      return len(self.qList)
````

### Problem: (스스로) 덱(double-ended queue) 구현하기

😄 덱은 double-ended queue의 줄임말로 큐의 전단(front)과 후단(rear)에서 모두 삽입과 삭제가 가능한 큐를 말한다. 큐에서와 같이 중간에 삽입하거나 삭제하는 것은 허용하지 않는다.
<div align="center"><img src="https://haesunbyun.github.io/common/images/adt7.png" width="400" height="100" style="width:600px;"></div>


- 덱의 ADT는 다음과 같다.

데이터: 전단과 후단을 통한 접근을 하는 원소들의 모임

연산(동작):
  - addFront(e) : 원소 e를 덱의 맨 앞에 추가한다.
  - addRear(e) : 원소 e를 덱의 맨 뒤에 추가한다.
  - deleteFront() : 큐의 맨 앞의 요소를 삭제하고 반환한다.
  - deleteRear() : 큐의 맨 뒤의 요소를 삭제하고 반환한다.
  - getFront() : 큐의 맨 앞의 요소를 삭제하지 않고 반환한다.
  - getRear() : 큐의 맨 뒤의 요소를 삭제하지 않고 반환한다.
  - isEmpty() : 큐가 비어있으면 True 아니면 False 반환
  - size() : 큐 내의 모든 원소 개수를 반환한다.
  - `__str__()` : 큐 내의 모든 요소를 dQ<  >dQ 형태로 출력한다.

### Problem: [스택 활용]계산기◼◼◼◼◼
😄 주어진 연산식을 스택을 활용하여 결과가 나오도록 하시오.

'2+3*4'과 같이 피연산자 사이에 연산자가 들어가는 방법을 중위표기법이라고 한다. 그러나 컴퓨터를 통해 각 연산자의 우선순위대로 계산하려면 후위표기식으로 변환해서 계산해야 한다.
후위 표기식은  연산자가 피연산자 뒤에 있는 표기식이다. '2+3*4'을 후위표기식으로 표현하면 '234*+'로 표기한다.

후위표기식으로 변환하는 방법은 다음과 같다.

1. 숫자는 바로 변환하는 후위식에 넣어준다.
2. 연산자일 경우 스택에 넣어준다. 이때 스택에 들어있는 연산자가 우선순위가 높거나 같다면 다 빼서 차례대로 후위식에 넣어준 후 스택에 현재 연산자를 넣는다.
3. 위 1과 2를 반복을 완료하고 스택에 남은 연산자를 모두 식에 넣어준다.

예를 들어, 2+3*4-9/3을 후위표기식으로 변환해보자.

|순서|설명|후위식|스택|
|:----------:|:----------|:----------|:----------|
|1|2는 숫자이므로 후위식에 넣어준다.|2|(비어있음)|
|2|+는 연산자이며, 스택이 비어있으므로 스택에 넣어준다.|2|+|
|3|3은 숫자이므로 후위식에 넣어준다.|23|+|
|4|*는 연산자이며, 스택에 있는 +는 우선순위가 낮으므로 스택에 넣어준다.|23|+*|
|5|4는 숫자이므로 후위식에 넣어준다.|234|+*|
|6|-는 연산자이며, 스택에 넣어준다.<br> 이때, 스택에 들어있는 연산자가 우선순위가 높기때문에 빼서 후위식에 넣어준다. <br>이후 -는 스택에 넣어준다.|234*+|-|
|7|9는 숫자이므로 후위식에 넣어준다.|234*+9|-|
|8|/는 연산자이며, 스택에 있는 연산자가 우선순위가 낮으므로 스택에 넣어준다.|234*+9|-/|
|9|3은 숫자이므로 후위식에 넣어준다.|234*+93|-/|
|10|더 이상 문자가 없으므로 스택에서 차례로 빼서 후위식에 넣어준다.|234*+93/-|(비어있음)|

후위식을 계산하는 방법은 다음과 같다.
1. 숫자이면 스택에 넣는다.
2. 연산자가 나오면 스택에 있는 것을 두 개 뺀 후 먼저 뺀 것을 뒤로 가게 해서 계산한 후 스택에 넣는다.

예를 들어, 234*+93/-를 계산해보자.
